# MATPLOTLIB VISUALIZATIONS


```python
rigion_sales = df.groupby("Region")["Sales"].sum()
```

```python
plt.figure(figsize=(8,5))
rigion_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
region_profit=df.groupby("Region")["Profit"].sum()
```

```python
plt.figure(figsize=(8,5))
region_profit.plot(kind="bar")
plt.title("profit by Region")
plt.xlabel("Region")
plt.ylabel("Total profit")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Sales"], df["Profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=5, edgecolor='black')

plt.title("Distribution of Sales", fontsize=16, fontweight='bold')
plt.xlabel("Sales ($)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Profit"], marker='o')
plt.title("Profit Trend")
plt.xlabel("Order Number")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(7,7))
rigion_sales.plot(kind="pie", autopct='%1.1f%%')
plt.title("Region Wise Sales Share")
plt.ylabel("")
plt.show()
```
